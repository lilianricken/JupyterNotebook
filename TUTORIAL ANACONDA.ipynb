```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
#carregando o arquivo
videoGame = pd.read_csv('vgsales.csv')
```

```python
#Renomeando colunas
videoGame.columns = ['Ranking', 'Nome', 'Plataforma', 'Ano',
                      'Gênero', 'Editora', 'Vendas América do Norte', 'Vendas EUA',
                      'Vendas Japão', 'Outras vendas', 'Vendas Global']
```

```python
videoGame['Plataforma'].value_counts()
```

```python
titulos_lancados = videoGame['Plataforma'].value_counts()
titulos_lancados.plot()
videoGame['Plataforma'].value_counts().plot()
```

```python
videoGame['Plataforma'].value_counts().head(10).plot(kind='bar', figsize=(11,5), grid=False, rot=0, color='green')

plt.title('Os 10 Videogames com mais titulos lançados')
plt.xlabel('Videogame')
plt.ylabel('Quantidade de jogos lançados')
plt.show()
```

```python
top10Vendidos = videoGame[['Nome', 'Vendas Global']].head(10).set_index('Nome').sort_values('Vendas Global', ascending = True)
top10Vendidos.plot(kind='barh', figsize=(11,7), grid = False, color = 'darkred', legend=False)

plt.title('Os 10 jogos mais vendidos no mundo')
plt.xlabel('Total de vendas (em milhões de dolares)')
plt.show()
```

```python
crosstab_vg = pd.crosstab(videoGame['Plataforma'], videoGame['Gênero'])
crosstab_vg.head()
```

```python
crosstab_vg['Total'] = crosstab_vg.sum(axis=1)
crosstab_vg.head()
```

```python
top10Plataforms = crosstab_vg[crosstab_vg['Total']>1000].sort_values('Total', ascending = False)
top10Final = top10Plataforms.append(pd.DataFrame(top10Plataforms.sum(), colums=['Total']).T, ignore_index=False)

sns.set(from_scale=1)
plt.figure(figsize=(18,9))
sns.heatmap(top10Final, annot=True, vmax=top10Final.loc[:'PS', :'Strategy'].values.max(), vmin=top10Final.loc[:, :'Strategy'].values.min(), fmt='d')
```
